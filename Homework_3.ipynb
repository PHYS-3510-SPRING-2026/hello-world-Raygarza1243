```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0, 20.01, 0.01)
t_plot = np.arange(-10, 10.01, 0.01)
def f(t):
    return np.sin(4 * t) * np.cos(4 * t)
def g(t):
    return np.where(t == 0, 1.0, np.sin(t) / t)
def h(t):
    return np.tan(2 * t)
```

```python
values = np.arange(-10, 10)
asymptotes = (2 * values + 1) * np.pi / 4
asymptotes = asymptotes[(asymptotes > -10) & (asymptotes < 10)]

ft = f(t_plot)
gt = g(t_plot)
ht = h(t_plot)
```

```python
plt.plot(t_plot, ft, label=r'f(t) = \sin(4t)\cos(4t)')
plt.plot(t_plot, gt, label=r'g(t) = \sin(t)/t')
plt.plot(t_plot, ht, label=r'h(t) = \tan(2t)')

for x_line in asymptotes:
    plt.axvline(x=x_line, color='gray', linestyle='--')

plt.title('f(t), g(t), and h(t)')
plt.legend(loc='upper right')
plt.xlabel('t (sec)')
plt.ylabel('Output Values')
plt.xlim(-10, 10)
plt.ylim(-5, 5)
plt.grid(True, which='both', linestyle=':')
plt.show()
```
