```python
def prime_num():
    prime_numbers = []
    for num in range(2,101):
        prime = True
        for i in range(2,num):
            if (num % i) == 0:
                prime = False
                break
        if prime:
            prime_numbers.append(num)

    return prime_numbers
primes = prime_num()
print("The prime numbers between 1 and 100 are:", primes)
```
