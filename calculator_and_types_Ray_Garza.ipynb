```python
# Q and N = number of integers being summed
# W and F = first integers in sequence
# E and L = last integers in sequence
# S = total sum
Q = 2
W = 1
E = 2
W + E == 3
3 * Q == 6
6 / 2 == 3

N = 2
F = 1
L = 2
S = (N * (F + L)) / 2
S == 3
if S == 3:
  print("The sum computed by brute force is",3.,
        "The sum computed by Gauss' formula is", S,
        "Both results are equal to each other.")
```
